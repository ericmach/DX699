# 2. Weekly graph question

Below are the graphs from last week's weekly graph question.  Suppose you wanted to cause the viewer to examine the last three months of data first.  How could you use preattentive attributes to do that?  Suggest several possibilities.

```python
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
numdates = 12
np.random.seed(0)
time_series = 100 * (np.random.normal(size = numdates) / 5 + np.arange(numdates) / 16)
datearray = pd.date_range(start='2023/01/01', end='2024/01/01', freq='M')
ax = sns.lineplot(x = datearray, y = time_series)
ax.set_xlabel("Year and month")
ax.set_ylabel("Nitrate level (% of eutrophic value)")
```

```python
np.random.seed(1)
time_series = 100 * (1 - (np.random.normal(size = numdates) / 8 + np.arange(numdates) / 16))
datearray = pd.date_range(start='2023/01/01', end='2024/01/01', freq='M')
ax = sns.lineplot(x = datearray, y = time_series)
ax.set_xlabel("Year and month")
ax.set_ylabel("Phosphate level (% of eutrophic value)")
```

To draw attention to the last three months of data first, a range of preattentive attributes could be utilized. For example, visual distinction could be drawn by shading the right area of the graph in the range of the last three months. In addition, the markers for the data points for the last three months could be made more distinct (X, O, *, etc.) and data labels could be added directly to those points indicating their phosphate/nitrate level explicitly. Alternatively, the last three months of data could be made a different color compared to the rest of the line graph, as this visual separation would naturally draw audience attention.

```python
import matplotlib.pyplot as plt
```

```python
# Figure 4.8, pp.111

categories = [f'Concern {i}' for i in range(1, 11)] 
values = (8.2, 8.6, 8.8, 10, 10.3, 11, 11.6, 11.6, 12.3, 12.9)

colors = ['red' if i > 2 else 'grey' for i in range(10)]

fig, ax = plt.subplots(figsize=(8, 6))
ax.barh(categories, values, color=colors)

ax.set_xlabel("Number of Concerns")
ax.set_ylabel("Concern")
ax.set_title("Top 10 Design Concerns")

plt.show()
```
